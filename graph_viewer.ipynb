```python
from IPython.display import Image
import json
import os

workspace_dir = '/home/saheli/work_collection'
json_file = os.path.join(workspace_dir, 'data_axs.json')

def process_json(json_file):
    with open(json_file, 'r') as file:
        json_data = json.load(file)
        if 'contained_entries' in json_data and isinstance(json_data['contained_entries'], dict):
            # convert keys to list and get last key , assuming the latest entry is a graph entry
            latest_key = list(json_data['contained_entries'].keys())[-1] 
            return latest_key
        else:
            print("No contained_entries found")

image_file_path = os.path.join(workspace_dir, process_json(json_file), 'image.png')
print('The image file path is at:', image_file_path)

Image(filename=image_file_path)
```
